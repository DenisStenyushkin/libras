```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
raw_df = pd.read_csv('movement_libras.data', header=None)
```

```python
raw_df.head()
```

```python
points_df = pd.DataFrame()
```

```python
points_df['absc'] = raw_df[list(i for i in range(0, 90, 2))].values.tolist()
points_df['ord'] = raw_df[list(i for i in range(1, 90, 2))].values.tolist()
points_df['class'] = raw_df[90]
```

```python
points_df.head()
```

```python
def plot_movement(row):
    x = np.array(row.iloc[0]['absc'])
    y = np.array(row.iloc[0]['ord'])

    plt.figure()
    plt.gcf().set_size_inches(10, 10)
    plt.quiver(x[:-1], y[:-1], x[1:]-x[:-1], y[1:]-y[:-1], scale_units='xy', angles='xy', scale=1, width=0.003)

    plt.show()
```

```python
plot_movement(points_df[points_df['class']==1].sample(n=1))
```

```python
plot_movement(points_df[points_df['class']==2].sample(n=1))
```
